```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure(畫布/透過實體方法figure傳出fig)
fig = plt.figure()

# 2.在1.的figure上製作一個axes，而(1,1,1)表示只畫一個圖表
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y) #產生ticks & labels

# 4.顯示圖
plt.show()
```

```python
#標準寫法
import matplotlib.pyplot as plt

#製作figure畫布
fig = plt.figure()

ax_1 = fig.add_subplot(1,3,1) #傳出一個畫布
ax_1.set_title("axes1") #畫布的title

ax_2 = fig.add_subplot(1,3,2) #傳出一個畫布
ax_2.set_title("axes2") #畫布的title

ax_3 = fig.add_subplot(1,3,3) #傳出一個畫布
ax_3.set_title("axes3") #畫布的title
```

```python
import matplotlib.pyplot as plt
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':90, 'english':75, 'math':98, 'history':65, 'discover':75}

#dictionary有一個功能是實體方法叫keys，用來輸出keys
student1.keys() #此type為like-list，因此可以用list方式來操作
subjects = list(student1.keys()) #直接轉成list
value1 = list(student1.values()) #一樣透過list將like-list的student1.values()轉成list
value2 = list(student2.values())
(subjects, value1, value2) #這是一個tuple

#建立一個畫布，畫布比例大小設為寬度8.0，高度5.0 / 畫布顏色設定時，顏色的表示法都是用#，RGB的6位數表示法，全部的顏色相同就是表示灰色
figure = plt.figure(figsize=(8.0, 5.0), dpi=250.0, facecolor="#cccccc")
type(figure) #傳出來是type figure / <Figure size 2000x1250 with 0 Axes>，但只有畫布並沒有圖表axes
ax = figure.add_subplot()
ax.plot(subjects, value1,'b-o') #查matplotlib.axes.Axes.plot可看見各參數定義
ax.plot(subjects, value2,'r--+')
plt.show()
```
